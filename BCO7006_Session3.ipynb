# Part 1 — Operators (≈ 20 min)

An **operator** is a symbol that does something to one or two values: `+`, `-`, `>`, `==`, `and`. We'll cover three families: arithmetic, comparison, and logical.

## 1.1 Arithmetic operators

The five you'll use constantly:

| Operator | Meaning | Example |
|---|---|---|
| `+` | Add | `10 + 3` → `13` |
| `-` | Subtract | `10 - 3` → `7` |
| `*` | Multiply | `10 * 3` → `30` |
| `/` | Divide (always returns a float) | `10 / 3` → `3.333...` |
| `**` | Power | `10 ** 3` → `1000` |

Two more worth knowing:

| Operator | Meaning | Example |
|---|---|---|
| `//` | Integer division (drop the remainder) | `10 // 3` → `3` |
| `%` | Remainder (modulo) | `10 % 3` → `1` |

````python
# A typical BA calculation: price * quantity, plus 10% tax
price = 49.99
quantity = 3
tax_rate = 0.10

subtotal = price * quantity
tax = subtotal * tax_rate
total = subtotal + tax

print("Subtotal:", subtotal)
print("Tax:    ", tax)
print("Total:  ", total)
````

````python
# Round the total to 2 decimal places using round()
total_rounded = round(total, 2)
print("Total (rounded): $", total_rounded)
````

## 1.2 Comparison operators

These compare two values and return a `True` or `False` (a *boolean*).

| Operator | Meaning |
|---|---|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal |
| `<=` | Less than or equal |

Note: `=` is assignment, `==` is comparison. Mixing them up is the most common beginner error in any language.

````python
order_total = 145.50

is_premium_order = order_total > 100
qualifies_free_shipping = order_total >= 75

print("Premium order? ", is_premium_order)
print("Free shipping? ", qualifies_free_shipping)
````

````python
# Comparisons work on text too — alphabetical ordering
print("apple" < "banana")   # True
print("apple" == "Apple")   # False — case matters!
print("apple" == "apple")   # True
````

# Part 2 — Lists (≈ 35 min)

A **list** is an ordered collection of items. You write it with square brackets `[ ]` and separate items with commas. It's the workhorse of Python — you'll use lists everywhere.

## 2.1 Creating a list

````python
# A list of product names
products = ["laptop", "mouse", "keyboard", "monitor", "webcam"]
print(products)

# Lists can hold numbers
prices = [999, 25, 75, 350, 80]
print(prices)

# They can even mix types (though you usually won't)
mixed = ["laptop", 999, True]
print(mixed)

# Empty list — useful when you want to add things to it later
empty = []
print(empty)
````

## 2.2 Accessing items: indexing and slicing

Lists are **zero-indexed**: the first item is at position `0`, the second at `1`, and so on. You can also count from the end with negative numbers.

````python
products = ["laptop", "mouse", "keyboard", "monitor", "webcam"]

print(products[0])    # First item
print(products[1])    # Second item
print(products[-1])   # LAST item (negative counts from the end)
print(products[-2])   # Second-to-last
````

````python
products = ["laptop", "mouse", "keyboard", "monitor", "webcam"]
products[ 2:3 ]
````

````python
# Slicing: get a SUB-list with [start:stop]
# Note: 'stop' is exclusive — it stops BEFORE that index
print(products[0:3])   # First three: indexes 0, 1, 2
print(products[1:4])   # Items at index 1, 2, 3
print(products[2:])    # From index 2 to the end
print(products[:2])    # From start up to (but not including) index 2
````

````python
print(products [:4]) #this is if i don't want to name the start it automatically assumes to start from 0
print(products[:
               ])
````

## 2.3 How long is the list? Is X in it?

````python
products = ["laptop", "mouse", "keyboard", "monitor", "webcam"]

print(len(products))             # How many items?
print("laptop" in products)      # Is "laptop" present?
print("printer" in products)     # Is "printer" present?
````

## 2.4 Modifying a list

Lists are **mutable** — you can change them after creation.

````python
products = ["laptop", "mouse", "keyboard"]

# Add to the end
products.append("monitor")
print(products)

# Insert at a specific position
products.insert(1, "headphones")    # Insert at index 1
print(products)

# Remove a specific item by value
products.remove("mouse")
print(products)

# Remove and return the last item
last = products.pop()
print("Removed:", last)
print(products)

# Replace an item by index
products[0] = "gaming laptop"
print(products)
````

# Part 3 — Strings as sequences (≈ 15 min)

Here's a useful idea: **a string is just a sequence of characters**, and a lot of what you learned about lists works on strings too.

````python
email = "alice@example.com"

print(email[0])       # First character
print(email[-1])      # Last character
print(email[0:5])     # Slice — first 5 chars
print(len(email))     # Length
print("@" in email)   # Membership check
````

## 3.1 Useful string methods

Strings come with built-in methods (functions you call with `.method_name()`).

````python
name = "  Alice Wong  "

print(name.strip())          # Remove whitespace from both ends
print(name.upper())          # ALL CAPS
print(name.lower())          # all lower case
print(name.replace("Alice", "Bob"))   # Replace text
````

````python
# .split() turns a string into a LIST — incredibly useful
email = "alice@example.com"
parts = email.split("@")
print(parts)
print("Username:", parts[0])
print("Domain:  ", parts[1])
````

````python
# The reverse: join a list back into a string
words = ["business", "analytics", "is", "fun"]
sentence = " ".join(words)    # The string before .join() is the separator
print(sentence)
````

## 3.2 f-strings: putting variables inside text

The cleanest way to build text that contains variables. Prefix the string with `f` and put variables in `{ }`.

````python
name = "Alice"
order_count = 7
total_spent = 432.50

# The old way (works but messy)
print("Customer " + name + " has " + str(order_count) + " orders.")

# f-string — much cleaner
print(f"Customer {name} has {order_count} orders, total ${total_spent}")

# You can do calculations inside the braces too
print(f"Average per order: ${total_spent / order_count:.2f}")
````

# Part 4 — Dictionaries (≈ 35 min)

A **dictionary** stores **key → value** pairs. Where a list is good for "an ordered bunch of things", a dictionary is good for "a labelled record" — like a row in a spreadsheet.

Syntax: curly braces `{ }`, with `key: value` pairs separated by commas.

## 4.1 Creating a dictionary

````python
# A single product as a dictionary
product = {
    "name": "laptop",
    "price": 999,
    "stock": 15,
    "category": "electronics"
}

print(product)
````

## 4.2 Accessing values by key

````python
product = {
    "name": "laptop",
    "price": 999,
    "stock": 15,
    "category": "electronics"
}

print(product["name"])      # 999
print(product["price"])     # 999

# WARNING: asking for a key that doesn't exist crashes your code
# Uncomment the next line to see the error
# print(product["color"])
````

## 4.3 Safe access with `.get()`

`.get()` returns `None` (or a default you specify) instead of crashing.

````python
print(product.get("color"))                 # None — no crash
print(product.get("color", "not specified"))  # Default value
print(product.get("price", 0))               # Returns the actual price (999)
````

## 4.4 Adding, updating, and deleting

````python
product = {
    "name": "laptop",
    "price": 999,
    "stock": 15
}

# Add a new key-value pair
product["color"] = "silver"
print(product)

# Update an existing value (same syntax)
product["price"] = 899
print(product)

# Delete a key
del product["stock"]
print(product)

# Check if a key exists
print("price" in product)
print("stock" in product)
````

## 4.5 Iterating over a dictionary — the recipe

You'll learn `for` loops properly next session. For now, here's a recipe to copy when you want to look at every entry in a dictionary:

```python
for key, value in some_dict.items():
    print(key, "->", value)
```

````python
product = {"name": "laptop", "price": 999, "stock": 15, "color": "silver"}

# The recipe in action
for key, value in product.items():
    print(f"{key}: {value}")
````

## 4.6 A more useful pattern: catalogue as a dictionary

Instead of one product, store many — using product names as keys and prices as values.

````python
catalogue = {
    "laptop":    999,
    "mouse":     25,
    "keyboard":  75,
    "monitor":   350,
    "webcam":    80
}

# Look up a price by product name — INSTANTLY
print("Mouse costs:", catalogue["mouse"])

# How many products?
print("Catalogue has", len(catalogue), "products")

# Loop through and show all
for name, price in catalogue.items():
    print(f"  {name:<10} ${price}")
````

# Part 5 — Tuples and Sets (≈ 15 min)

Two more containers. We'll touch them lightly — you'll meet them more deeply when you need them.

## 5.1 Tuples — a list that can't change

A **tuple** is like a list, but **immutable** — once created, you can't modify it. Use parentheses `( )` instead of square brackets.

When do you want immutability? When the data is conceptually a fixed grouping that shouldn't change: coordinates, an RGB colour, a (latitude, longitude) pair.

````python
# A geographic point
melbourne = (-37.8136, 144.9631)
print(melbourne)
print("Latitude:", melbourne[0])
print("Longitude:", melbourne[1])

# Tuples support the same indexing as lists
# But this would crash:
# melbourne[0] = 0   # TypeError: 'tuple' object does not support item assignment
````

````python
# A handy tuple trick: returning multiple values at once
def get_product_info():
    return ("laptop", 999, 15)   # returns a tuple

# Unpack into separate variables
name, price, stock = get_product_info()
print(f"{name} — ${price} ({stock} in stock)")
````

## 5.2 Sets — unique items, fast membership

A **set** holds unique items, unordered, with very fast "is X in this collection?" checks. Great for things like "the set of customers who bought product X".

````python
# Curly braces (like dicts) but no key:value — just values
customers_jan = {"alice", "bob", "carol", "dave"}
customers_feb = {"bob", "carol", "eve", "frank"}

# Notice duplicates get removed automatically
duplicates_removed = {"a", "b", "a", "c", "b"}
print(duplicates_removed)
````

````python
# The killer feature: set operations
customers_jan = {"alice", "bob", "carol", "dave"}
customers_feb = {"bob", "carol", "eve", "frank"}

# Customers who bought in BOTH months (intersection)
loyal_customers = customers_jan & customers_feb
print("Loyal:", loyal_customers)

# Customers in EITHER month (union)
all_customers = customers_jan | customers_feb
print("Total unique customers:", len(all_customers))

# Customers in Jan but NOT Feb (difference) — at risk of churn
churned = customers_jan - customers_feb
print("Churned:", churned)
````
